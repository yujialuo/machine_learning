## Q1. Kmeans

```python
mnist_data = sio.loadmat("hw7_data/mnist_data/images.mat")
mnist_x = mnist_data['images']
mnist_x = np.transpose(mnist_x, (2, 0, 1))
mnist_x = np.reshape(mnist_x, (60000, 28*28))
```

```python
def distance(x, y):
    result = np.linalg.norm(x * y)**2
    return result

class Kmeans:
    def __init__(self, k, x):
        self.x = x
        self.n = len(x)
        self.k = k
        self.mean = [0] * self.k
        self.labels = np.array([randint(0, k-1) for _ in range(self.n)])

    def train(self, max_step):

        prev_labels = np.zeros(self.n)
        count = 0
        while not np.array_equal(self.labels, prev_labels):
#             print(count)
            prev_labels = np.copy(self.labels)
            # Update mean
            for y in range(self.k):
                indices = np.where(self.labels == y)[0]
                if list(indices):
                    self.mean[y] = np.mean(self.x[indices], axis=0)

            # Update label
            for i in range(self.n):
                new = self.find_closest_label(self.x[i])
                self.labels[i] = new
            
            if count == max_step:
                return
            count += 1

    def find_closest_label(self, x):
        a = []
        for i in range(self.k):
            mean = self.mean[i]
            a.append(distance(x, mean))
        return np.argmin(np.array(a))
```

```python
k = 5
k5 = Kmeans(k, mnist_x)
k5.train(100)
```

```python
k = 10
k10 = Kmeans(k, mnist_x)
k10.train(100)
```

```python
k = 20
k20 = Kmeans(k, mnist_x)
k20.train(100)
```

## Q3. Joke Recommendation

### 3.1 Data Preprocessing

```python
train = sio.loadmat("hw7_data/joke_data/joke_train.mat")['train']
train_clean = np.nan_to_num(train)
```

```python
val = []
f = open("hw7_data/joke_data/validation.txt", "r")
lines = f.readlines()
for line in lines:
    line = line.strip().split(',')
    a = [int(line[0]), int(line[1]), int(line[2])] # user, joke, label
    val.append(a)
val = np.array(val)
```

### 3.2 Model

#### a) SVD

```python
def PCA(x, n=0):
    u, s, v = linalg.svd(x, full_matrices=False)
    sig = np.copy(s)
    sig[n:] = 0
    temp = np.matmul(u, np.diag(sig))
    f = np.matmul(temp, v)
    return f

def mse(a, b):
    a = np.reshape(a, (a.shape[0]*a.shape[1],))
    b = np.reshape(b, (b.shape[0]*b.shape[1],))
    result = np.mean((a-b) ** 2)
    return result

def validate(val, pred):
    correct = 0
    for line in val:
        i = line[0] - 1
        j = line[1] - 1
        y = line[2]
        p = pred[i][j]
        if int(y) == int(p):
            correct += 1
    acc = float(correct) / len(val)
    return acc

def transform(y):
    new = [[0] * len(y[0])] * len(y)
    for i in range(len(y)):
        for j in range(len(y[0])):
            if y[i][j] > 0:
                new[i][j] = 1
            else:
                new[i][j] = 0
    return new
```

```python
# d = 2, 5, 10, 20
for d in [2, 5, 10, 20]:
    pred = PCA(train_clean, d)
    m = mse(pred, train_clean)
    pred = transform(pred)
    acc = validate(val, pred)
    print("d = {}, mse = {}, accuracy = {}".format(d, m, acc))
```

```python
# def update(x, i, j, u, v, lam):
#     while True:
#         vj = v[:, j]
#         rij = x[i][j]
#         ui = np.matmul(np.linalg.inv(np.outer(vj, vj) + np.diag(np.full((len(vj), ), lam))), (rij * vj))
#         if (np.allclose(ui, u[i], rtol = 1e-02, atol = 1e-04)):
#             break
#         u[i] = ui
#         vj = np.matmul(np.linalg.inv(np.outer(ui, ui) + np.diag(np.full((len(ui), ), lam))), (rij * ui))
#         if np.allclose(vj, v[:,j], rtol = 1e-02, atol = 1e-04):
#             break
#         v[:, j] = vj
#     return u, v

def predict(x, origin_x, lam, d):
    u, s, v = linalg.svd(x, full_matrices=False)
    v = np.matmul(np.diag(s), v)
    u = u[:, :d]
    v = v[:d]
    for step in range(2):
        for i in range(u.shape[0]):
            for j in range(0, v.shape[1]):
                # update u
                inds = np.nonzero(x[i] != 0)
                inds = [x for x in inds]
                v_tilt = (v.T[inds]).T
                u[i] = np.matmul(np.matmul(np.linalg.inv(np.matmul(v_tilt, v_tilt.T) + lam * np.eye(d)), v), x[i])
                # update v
                inds = np.nonzero(x[:, j] == 0)
                u_tilt = u[inds[0]]
                v[:, j] = np.matmul(np.matmul(np.linalg.inv(np.matmul(u_tilt.T, u_tilt) + lam * np.eye(d)), u.T), x[:,j])
    y = np.matmul(u, v)
    return y
```
