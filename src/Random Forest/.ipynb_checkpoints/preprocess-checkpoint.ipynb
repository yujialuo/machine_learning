```python
import numpy as np
import pandas as pd
import csv
import warnings
warnings.filterwarnings('ignore')
```

## Census

```python
df = pd.read_csv('census/train_data.csv') # this creates a DataFrame 
missing = ['workclass', 'occupation', 'native-country']
for col in missing:
    mean = df[col].value_counts().index[0]
    df[col].replace('?', mean, inplace=True)
df = pd.get_dummies(df) 
df.to_csv('census/census_clean.csv')
df
```

```python
df = pd.read_csv('census/test_data.csv') # this creates a DataFrame 
missing = ['workclass', 'occupation', 'native-country']
for col in missing:
    mean = df[col].value_counts().index[0]
    df[col].replace('?', mean, inplace=True)
df = pd.get_dummies(df) 
df.to_csv('census/census_test_clean.csv')
```

## Titanic

```python
df = pd.read_csv('titanic/titanic_training.csv')
df = df.fillna(df.mean())
df['cabin'] = df['cabin'].str[:3]
df['ticket'] = df['ticket'].str.extract('(\d+)').astype(float)
df['cabin_num'] = df['cabin'].str.extract('(\d+)').astype(float)
df['cabin_letter'] = df['cabin'].str[0]

mean = df['cabin_num'].value_counts().index[0]
df['cabin_num'].fillna(mean, inplace=True)
mean = df['cabin_letter'].value_counts().index[0]
df['cabin_letter'].fillna(mean, inplace=True)
mean = df['ticket'].value_counts().index[0]
df['ticket'].fillna(mean, inplace=True)

del df['cabin']
df = pd.get_dummies(df) 
df.to_csv('titanic/titanic_clean.csv')
df
```

```python
df = pd.read_csv('titanic/titanic_testing_data.csv')
df = df.fillna(df.mean())
df['cabin'] = df['cabin'].str[:3]
df['ticket'] = df['ticket'].str.extract('(\d+)').astype(float)
df['cabin_num'] = df['cabin'].str.extract('(\d+)').astype(float)
df['cabin_letter'] = df['cabin'].str[0]

mean = df['cabin_num'].value_counts().index[0]
df['cabin_num'].fillna(mean, inplace=True)
mean = df['cabin_letter'].value_counts().index[0]
df['cabin_letter'].fillna(mean, inplace=True)
mean = df['ticket'].value_counts().index[0]
df['ticket'].fillna(mean, inplace=True)

del df['cabin']
df = pd.get_dummies(df) 
df.to_csv('titanic/titanic_test_clean.csv')
```
