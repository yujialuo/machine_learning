```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import operator
import pandas as pd

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
idk = [1-0.7511, 1-0.6963, 1-0.5900, 1-0.4558, 1-0.3926, 1-0.3474, 1-0.2111]
# alex = [0.7053]
tp_rate = [0.7589,0.8216, 0.8457,0.8349, 0.7976, 0.7758, 0.6683]
acc = np.subtract(1,idk) * tp_rate + idk
fig, ax = plt.subplots()

ax.axhline(y=1.0, color="red", linestyle='dotted', label="faster-rcnn acc")
ax.scatter(idk, acc, color="green")
ax.plot(idk, acc, label='cascade acc', color="green")

# ax.scatter(idk, alex, color="blue")
# ax.plot(idk, alex, label='alexnet acc', color="blue")

ax.scatter(idk, tp_rate, color="b")
ax.plot(idk, tp_rate, label='alexnet exit samples acc', color="b")


ax.set_xlabel('IDK Rate')
ax.set_ylabel('Test Accuracy')
ax.set_title("Hard Distilation Cascade")
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
             box.width, box.height * 0.9])
lgd = ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.ylim((0.0, 1.05))
plt.xlim((-0.01, 1.01))
fig.savefig("plot.pdf", bbox_extra_artists=(lgd,), bbox_inches='tight')
```

```python
def load_data(file):
    with open(file, 'rb') as fp:
        d = pickle.load(fp, encoding='latin1')
    return d

def concact(data_lst):
    data = {}
    data['cascade_accs'], data['idk_rates'] = [], []
    for dy in data_lst:
        data['cascade_accs'] += dy['cascade_accs']
        data['idk_rates'] += dy['idk_rates']

    # Reorder
    lst = zip(data['idk_rates'], data['cascade_accs'])
    lst = sorted(lst, key=lambda x: x[0])
    data['idk_rates'] = [x[0] for x in lst]
    data['cascade_accs'] = [x[1] for x in lst]
    
    return data

def reorder(data):
    lst = zip(data['idk_rates'], data['cascade_accs'], data['lam'], data['flip_coin'])
    lst = sorted(lst, key=lambda x: x[0])
    data['idk_rates'] = [x[0] for x in lst]
    data['cascade_accs'] = [x[1] for x in lst]
    data['lam'] = [x[2] for x in lst]
    data['flip_coin'] = [x[3] for x in lst]
    return data

def plot(data, title, path):
    fig, ax = plt.subplots()

    ax.axhline(y=0.336, color="black", linestyle='dotted', label="simple")
    ax.axhline(y=0.80, color="grey", linestyle='dotted', label="complex")
    ax.scatter(data['idk_rates'], data['cascade_accs'], color="red")
    ax.plot(data['idk_rates'], data['cascade_accs'], label='cascade', color="red")
    ax.scatter(data['idk_rates'], data['flip_coin'], color="green")
    ax.plot(data['idk_rates'], data['flip_coin'], label='flip-coin', color="green")

    ax.set_xlabel('IDK Rate')
    ax.set_ylabel('Test Accuracy')
    ax.set_title(title)
    box = ax.get_position()
    ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 0.9])
    lgd = ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    plt.ylim((0.25, 0.9))
    plt.xlim((-0.01, 1.01))
    fig.savefig(path + "plot.pdf", bbox_extra_artists=(lgd,), bbox_inches='tight')
```

# Dynamic Cascade
* 11-Class Simple Model 

#### Masked
Trained 150K batches, Initial Ir = 1e-3, decay exponentially w/ decay_step 5000 and decay_rate 0.1

```python
data = {'idk_rates': [1, 0.09395, 0.23688, 0.15755, 0.613, 0.087, 0.1248, 0.6284],
       'cascade_accs': [0.8, 0.38,0.451, 0.409, 0.622, 0.38, 0.396, 0.6262],
       'lam': [1, 50, 15, 20, 5, 30, 25, 3]}
assert len(data['idk_rates']) == len(data['lam']) == len(data['cascade_accs'])
data['flip_coin'] = [x*0.795 + (1-x)*0.336 for x in data['idk_rates']]
data = reorder(data)
path = 'result/dynamic/with-mask/'
title = 'Cifar-10 Dynamic Cascade (Masked)'
plot(data, title, path)
pd.DataFrame(data)
```

$Loss_{dy, mask} = \frac{1}{n} \Sigma (1 - P_{idk}) \cdot L_{s} + \lambda \cdot I(\hat{y_{s}} = idk) \cdot P_{idk} \cdot (L_{c} + 0.5) $

#### No Mask

```python
data = {'idk_rates': [0.008914, 0.62, 0.65184, 1.0],
       'cascade_accs': [0.33153,0.6264, 0.635, 0.8],
       'lam': [3, 0.8, 0.5, 0.0001]}
assert len(data['idk_rates']) == len(data['lam']) == len(data['cascade_accs'])
data['flip_coin'] = [x*0.8 + (1-x)*0.336 for x in data['idk_rates']]
data = reorder(data)
path = 'result/dynamic/no-mask/'
title = 'Cifar-10 Dynamic Cascade (No Mask)'
plot(data, title, path)
pd.DataFrame(data)
```

$Loss_{dy} = \frac{1}{n} \Sigma (1 - P_{idk}) \cdot L_{s} + \lambda \cdot P_{idk} \cdot (L_{c} + 0.5) $

# Entropy Cascade
* Only has a 10-class simple model, use a threshold to determine IDK

```python
data = {'idk_rates': [0.17],
       'cascade_accs': [0.32],
       'lam': [200]}
assert len(data['idk_rates']) == len(data['lam']) == len(data['cascade_accs'])
data['flip_coin'] = [x*0.8 + (1-x)*0.336 for x in data['idk_rates']]
data = reorder(data)
path = 'result/entropy/with-mask/'
title = 'Cifar-10 Entropy Cascade (Masked)'
plot(data, title, path)
pd.DataFrame(data)
```

$Loss_{en, mask} = \frac{1}{n} \Sigma L_{s} + \lambda \cdot I(\hat{y_{s}} < thr) \cdot (L_{c} + 0.5) $

```python
data = {'idk_rates': [0.99, 0.9676],
       'cascade_accs': [0.797, 0.7857],
       'lam': [180, 220]}
assert len(data['idk_rates']) == len(data['lam']) == len(data['cascade_accs'])
data['flip_coin'] = [x*0.8 + (1-x)*0.336 for x in data['idk_rates']]
data = reorder(data)
path = 'result/entropy/with-mask/'
title = 'Cifar-10 Entropy Cascade (No-Mask)'
plot(data, title, path)
pd.DataFrame(data)
```

$Loss_{en} = \frac{1}{n} \Sigma L_{s} + \lambda \cdot (L_{c} + 0.5) $
